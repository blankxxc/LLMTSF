```python
def mean_squared_error(list1, list2):
    if len(list1) != len(list2):
        raise ValueError("Lists must be of the same length.")

    squared_error = [(x - y) ** 2 for x, y in zip(list1, list2)]
    mse = sum(squared_error) / len(list1)
    return mse

# Example lists
list1 = list(sample1_Y)
list2 = [86.38775067809804,86.9454424380208,87.21348665399003,87.33091922042868,87.13090032387464,87.51225688657816,87.34840036654855,88.37502271430206,89.60430892403369,89.29098465938816,90.07243713639411,90.15051621168425,90.33454378092472,91.48792612557061]

# Calculate MSE
mse = mean_squared_error(list1, list2)
print("Mean Squared Error:", mse)
```

```python
import numpy as np
import pandas as pd
data=pd.read_csv('./data/AAPL_process.csv')
data = data['Open']

def create_sliding_windows(series, window_size, forecast_step):
    X, y = [], []
    for i in range(len(series) - window_size - forecast_step + 1):
        # 选择当前窗口的数据
        window = series[i:(i + window_size)]
        # 选择预测目标
        target = series[(i + window_size):(i + window_size + forecast_step)]
        X.append(window)
        y.append(target)
    return np.array(X), np.array(y)

# 假设窗口大小为24（例如，使用过去24个时间点的数据来预测）
window_size = 96
forecast_step = 96 # 预测步长为96

X, y = create_sliding_windows(data, window_size, forecast_step)

# 显示结果的维度以确保正确
X.shape, y.shape
import re

# 使用正则表达式将一个或多个空格替换为一个逗号

ts=str(X[0]).replace("[", "").replace("]", "").replace("\n", "")
ts = re.sub(r'\s+', ',', ts)
ts
```
