```python
import pandas as pd
import numpy as np
file_path = './data/AAPL_process.csv'
data = pd.read_csv(file_path)
# 设置滑动窗口大小和预测长度
window_size = 480
predict_length = 96

# 仅提取开盘价（Open）作为特征
open_prices = data['Open'].values

# 初始化特征和标签列表
features = []
labels = []

# 使用滑动窗口方法生成特征和标签
for i in range(len(open_prices) - window_size - predict_length):
    features.append(open_prices[i:i+window_size])
    labels.append(open_prices[i+window_size:i+window_size+predict_length])
train=features
test=labels
```

```python
def mean_squared_error(y_true, y_pred):
    """
    计算均方误差(MSE)
    
    参数:
    y_true -- 实际值，列表或数组。
    y_pred -- 预测值，列表或数组。
    
    返回:
    mse -- 均方误差值。
    """
    
    # 计算MSE
    mse = sum((y_true_i - y_pred_i) ** 2 for y_true_i, y_pred_i in zip(y_true, y_pred)) / len(y_true)
    return mse
all_mse=0
for i in range(len(test)):
    # 示例数据
    y_true = np.array(test[i])  # 实际值
    y_pred = np.array(samples_data[i])
    # y_pred = np.array(pd.DataFrame(samples_data[i]).median())
    # 计算MSE
    mse = mean_squared_error(y_true, y_pred)
    all_mse+=mse
print(f"MSE: {all_mse/len(test)}")
```
